```python
### FINAL PROJECT ###

# The main function #

######Set up and check GPU########
import tensorflow as tf
# device_name = tf.test.gpu_device_name()
# if device_name != '/device:GPU:0':
#   raise SystemError('GPU device not found')
# print('Found GPU at: {}'.format(device_name))


####Set Up Keras######
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.utils import plot_model
# import graphviz
import numpy
import pandas as pd

# set the seed for reproducibility
#numpy.random.seed(7)

#######Load Data as DF#######  
# dataset = numpy.loadtxt("poker_data_clean.csv", delimiter=",",skiprows=1)
dataset = pd.read_csv("poker_data_clean_2.csv")
dataset = dataset.sample(frac=1).reset_index(drop=True)
# SHUFFLE the rows
# list(dataset)
# dataset
```

```python
# from main import *
import io

# Create training set
# split into input (X) and output (Y) variables
X = dataset.iloc[0:80000,7:73]
Y = dataset.iloc[0:80000,73:79]
# Create test set
Xtest = dataset.iloc[80000:90000,7:73]
Ytest = dataset.iloc[80000:90000,73:79]


"""# Section 1: Set up the Model"""
for i in range(1):
    
    # First create Model for all of our data
    model = Sequential()
    #model.add(Dense(100,input_dim=66, activation='relu'))
    #model.add(Dropout(0.5))
    #model.add(Dense(i, input_dim=66, activation='relu'))
    #model.add(Dropout(0.50))
    model.add(Dense(48,input_dim = 66,activation='relu'))
    model.add(Dropout(0.25))
    model.add(Dense(20, activation='relu'))
    model.add(Dropout(0.25))
    model.add(Dense(6,activation='softmax'))
    # Compile model
    sgd = keras.optimizers.SGD(lr=0.01, momentum=0.8, decay=0.0001, nesterov=True)
    # adam = keras.optimizers.Adam(lr=0.020, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.00001, amsgrad=False)
    model.compile(loss='categorical_crossentropy', optimizer= sgd, metrics=['categorical_accuracy'])

    # Fit the model
    model.fit(X, Y, epochs= 250, batch_size=32)

    # b = model.predict_classes(x= X, batch_size= 32)
    #  1 Hidden Layer -> 87.56% acc 66 -> 15 -> 6 for 250 epochs
    #  1 Hidden Layers -> 88.10% acc 66 -> 30 -> 6 for 150 epochs
    # evaluate the model
    scores = model.evaluate(Xtest, Ytest)
    print("\n%.0f : %s: %.2f%%" % (i,model.metrics_names[1], scores[1]*100))
```
